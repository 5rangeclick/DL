```python
import pandas as pd

# 엑셀 파일 불러오기
file_path = '/content/통test.xlsx'
df = pd.read_excel(file_path)

# 첫 번째 열의 '\n'을 띄어쓰기로 바꾸기
df['news'] = df['news'].str.replace('\n', ' ')

# 변경된 데이터프레임을 다시 엑셀 파일로 저장
df.to_excel(file_path, index=False)  # 저장할 때 index를 포함하지 않으려면 index=False

# 변경된 데이터프레임 확인
print(df)
```
